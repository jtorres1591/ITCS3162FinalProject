```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon_df = pd.read_csv('pokemon.csv')
# Reading the file.
```

```python
pokemon_df
# Ensuring the file works.
```

```python
pokemon_df.info()
# Check data types.
```

```python
pokemon_df.isna().sum()
# Find null values.
```

```python
pokemon_df['type2'] = pokemon_df['type2'].fillna('None')
# Since there are null values in Type 2, which should be blank for many Pokemon, it will be replaced with None to avoid issues with null values.
```

```python
pokemon_df
# Ensuring the single types have type2 as None.
```

```python
# Creating and showing the correlation matrix of the dataset's features.
corr_matrix = pokemon_df.corr()
plt.figure(figsize=(20, 18))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
